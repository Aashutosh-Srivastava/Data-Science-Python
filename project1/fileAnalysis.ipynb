````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv('Expanded_data_with_more_features.csv')
df.head()
````

````python
df.describe()
````

````python
df.info()
````

````python
df.isnull().sum()
````

````python
#dropping unnamed column
df = df.drop('Unnamed: 0',axis=1)
df.head()
````

````python
#Gender Distribution
plt.figure(figsize=(6,5))
ax = sns.countplot(x='Gender',data=df)
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
````

```From the above chart we have analyzed that the number of females in tha data is more than the number of males```

````python
gb = df.groupby('ParentEduc').agg({'MathScore':'mean',
                                  'ReadingScore':'mean',
                                  'WritingScore':'mean'})
gb
````

````python
sns.heatmap(gb,annot = True)
plt.title('Relationship between parents education and student score')
plt.show()
````

```From the above the chart we have concluded that the parents education have a good impact on students scores```

````python
gb1 = df.groupby('ParentMaritalStatus').agg({'MathScore':'mean',
                                  'ReadingScore':'mean',
                                  'WritingScore':'mean'})
gb1
````

````python
sns.heatmap(gb1,annot = True)
plt.title('Relationship between parents marital status and student score')
plt.show()
````

```From the above chart we have concluded that there is no/negligible impact on students score due to their parents marital status ```

### Distribution of Ethinic Groups

````python
GroupA = df.loc[(df['EthnicGroup']=='group A')].count()
GroupB = df.loc[(df['EthnicGroup']=='group B')].count()
GroupC = df.loc[(df['EthnicGroup']=='group C')].count()
GroupD = df.loc[(df['EthnicGroup']=='group D')].count()
GroupE = df.loc[(df['EthnicGroup']=='group E')].count()
mlist = [GroupA['EthnicGroup'],GroupB['EthnicGroup'],GroupC['EthnicGroup'],GroupD['EthnicGroup'],GroupE['EthnicGroup']]
l = ['group A','group B','group C','group D','group E']
plt.title("Distribution of ethnic groups")
plt.pie(mlist, labels=l,autopct ="%1.2f%%" )
plt.show()
````
